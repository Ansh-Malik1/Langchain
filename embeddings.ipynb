```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings= (
    OllamaEmbeddings(model='gemma:2b')  # by default it uses llama2
)
```

```python
r1 = embeddings.embed_documents([
    "hello everyone. This is test string 1",
    "This is text string 2 for experimental purposes."
])
```

```python
r1[0]
```

```python
len(r1[0])
```

```python
embeddings.embed_query('This is an example of embedding a query using ollama embeddings')
```

##### Hugging Face Embeddings


```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HUGGINGFACE_API'] = os.getenv('HUGGINGFACE_API')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings2 = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
text = "this is an example of embedding using hugging face"
result = embeddings2.embed_query(text)
result
```
